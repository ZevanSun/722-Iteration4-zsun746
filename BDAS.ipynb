```python
states["Marketplace Tax Credits Rate (2016)"] = states["Marketplace Tax Credits (2016)"]/states["Marketplace Health Insurance Coverage (2016)"]
print(states["Marketplace Tax Credits Rate (2016)"].head())
```

```python
states1 = states.drop("Medicare Enrollment (2016)",axis=1)
print(states1.dtypes)
```

```python
states2 = states["Medicare Enrollment (2016)"]
print(states2.head())
```

```python
states = pd.concat([states1, states2],axis=1)
print(states.dtypes)
```

```python
print(states[["Uninsured Rate (2010)","Uninsured Rate (2015)","Average Monthly Tax Credit (2016)"]].head())
print(states[["Uninsured Rate (2010)","Uninsured Rate (2015)","Average Monthly Tax Credit (2016)"]].dtypes)

for i in range(states.shape[0]):   
    states["Uninsured Rate (2010)"][i] = states["Uninsured Rate (2010)"][i][:-1]
    states["Uninsured Rate (2015)"][i] = states["Uninsured Rate (2015)"][i][:-1]
    states["Average Monthly Tax Credit (2016)"][i] = states["Average Monthly Tax Credit (2016)"][i][1:]

states["Uninsured Rate (2010)"] = pd.to_numeric(states["Uninsured Rate (2010)"])
states["Uninsured Rate (2015)"] = pd.to_numeric(states["Uninsured Rate (2015)"])
states["Average Monthly Tax Credit (2016)"] = pd.to_numeric(states["Average Monthly Tax Credit (2016)"])

print(states[["Uninsured Rate (2010)","Uninsured Rate (2015)","Average Monthly Tax Credit (2016)"]].head())
print(states[["Uninsured Rate (2010)","Uninsured Rate (2015)","Average Monthly Tax Credit (2016)"]].dtypes)
```

```python
print(states["State Medicaid Expansion (2016)"].head())
print(states["State Medicaid Expansion (2016)"].dtypes)
states["State Medicaid Expansion (2016)"] = pd.to_numeric(states["State Medicaid Expansion (2016)"])
print(states["State Medicaid Expansion (2016)"].head())
print(states["State Medicaid Expansion (2016)"].dtypes)
```

```python
states = states.drop(["State","Uninsured Rate (2010)","Uninsured Rate (2015)","Average Monthly Tax Credit (2016)",
                      "State Medicaid Expansion (2016)","Marketplace Tax Credits Rate (2016)"],axis=1)
print(states.dtypes)
```

```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('tree_methods_adv').getOrCreate()
```

```python
data = spark.read.csv('states_new.csv',inferSchema=True,header=True)
```

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```python
data.columns
```

```python
assembler = VectorAssembler(
  inputCols=['Health Insurance Coverage Change (2010-2015)',
             'Employer Health Insurance Coverage (2015)',
             'Marketplace Health Insurance Coverage (2016)',
             'Marketplace Tax Credits (2016)',
             'Medicaid Enrollment (2013)',
             'Medicaid Enrollment (2016)'],

              outputCol="features")
```

```python
output = assembler.transform(data)
```

```python
final_data = output.select("features",'Medicare Enrollment (2016)')
```

```python
train_data,test_data = final_data.randomSplit([0.7,0.3])
```

```python
from pyspark.ml.classification import DecisionTreeClassifier,GBTClassifier,RandomForestClassifier
from pyspark.ml import Pipeline

from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
rf = RandomForestRegressor(labelCol='Medicare Enrollment (2016)',featuresCol="features")
rf_model = rf.fit(train_data)
```

```python
predictions = rf_model.transform(test_data)
```

```python
evaluator = RegressionEvaluator(labelCol="Medicare Enrollment (2016)", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```
